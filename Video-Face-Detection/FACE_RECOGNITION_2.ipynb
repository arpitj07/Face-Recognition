```python
import cv2 
import face_recognition
```

```python
input_video= cv2.VideoCapture("input.mp4")
```

find the number of frames in the video file

```python
length = int(input_video.get(cv2.CAP_PROP_FRAME_COUNT))
```

```python
fourcc = cv2.VideoWriter_fourcc('M','P','E','G')
```

```python
output_video = cv2.VideoWriter("output.avi",fourcc,24.00,(640,360))
```

Load some sample images 

```python
female_image = face_recognition.load_image_file("anushka-sharma.jpeg")
female_face_encoding = face_recognition.face_encodings(female_image)
```

```python
male_image = face_recognition.load_image_file("shahrukh-khan.jpg")
male_face_encodings = face_recognition.face_encodings(male_image)
```

```python
known_faces = [female_face_encoding,male_face_encodings]
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img=mpimg.imread('anushka-sharma.jpeg')
imgplot = plt.imshow(img)
plt.show()
```

```python
image = mpimg.imread("shahrukh-khan.jpg")
imgplot = plt.imshow(image)
plt.show()
```

Initialize some variables 

```python
face_locations =[]
face_encodings=[]
face_names=[]
frame_number=0
```

```python
#Grab single frame of video
while True:
    ret , frame = input_video.read()
    frame_number +=1
    
    #Quit when Video Ends
    if not ret:
        break
    #convert image from BGR to RGB 
    rgb_frames = frame[:,:,::-1]
    
    # find all face encodings in the current frame
    face_locations = face_recognition.face_locations(rgb_frames)
    face_encodings = face_recognition.face_encodings(rgb_frames,face_locations)
    
    
    for face_encoding in face_encodings:
        match = face_recognition.compare_faces(known_faces,face_encoding,tolerance=0.40)
        name = None
        if match[0].any():
            name= "Anushaka"
        elif match[1].any():
            name ='SRK'
            
            
        face_names.append(name)    
    
    #label the results
    for (top,right,bottom,left), name in zip(face_locations,face_names):
        if not name:
            continue
            
        cv2.rectangle(frame,(left,top),(right,bottom),(0,0,255),2)
        
        cv2.rectangle(frame,(left,bottom-25),(right,bottom),(0,0,255),cv2.FILLED)
        font = cv2.FONT_HERSHEY_DUPLEX
        cv2.putText(frame,name,(left+6,bottom-6),font,0.5,(255,255,255),1)
        
        
        
    print("writing frame {}/{}".format(frame_number,length))
    output_video.write(frame)
```

```python
# All done!
input_video.release()
cv2.destroyAllWindows()
```
